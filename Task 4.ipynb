```python
import pandas as pd

file_path = r"C:\Users\shubh\Downloads\USvideos.csv"
df = pd.read_csv(file_path)

df.head()
```

```python
df.info()
df.describe()
```

```python
missing_values = df.isnull().sum()

import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
sns.heatmap(df.isnull(), cbar=False, cmap='viridis')
plt.title('Missing Data Heatmap')
plt.show()
```

```python
numeric_columns = df.select_dtypes(include=['int64', 'float64']).columns

plt.figure(figsize=(16, 10))
for i, col in enumerate(numeric_columns, 1):
    plt.subplot(3, 3, i)
    sns.histplot(df[col], kde=True, bins=30)
    plt.title(f'Distribution of {col}')
plt.tight_layout()
plt.show()
```

```python
columns_to_remove = ['thumbnail_link','description']
df=df.drop(columns=columns_to_remove)
df.info()
```

```python
from datetime import datetime
import datetime
```

```python
df["trending_date"] = df["trending_date"].apply(lambda x : datetime.datetime.strptime(x, '%y.%d.%m' ))
df.head(3)
```

```python
df["publish_time"] = pd.to_datetime(df["publish_time"])
df.head()
```

```python
df["publish_month"] = df["publish_time"].dt.month
df["publish_day"] = df["publish_time"].dt.day
df["publish_hour"] = df["publish_time"].dt.hour
df.head(2)
```

```python
print(sorted(df["category_id"].unique()))
[1,2,10,15,17,19,20,22,23,24,25,26,27,28,29,30,43]
```

```python
import numpy as np
df['category_name']=np.nan
df.loc[(df['category_id']==1),'category_name']="Film and Animation"
df.loc[(df['category_id']==2),'category_name']="Autos ands Vehicles"
df.loc[(df['category_id']==10),'category_name']="Music"
df.loc[(df['category_id']==15),'category_name']="Pets and Animals"
df.loc[(df['category_id']==17),'category_name']="Sports"
df.loc[(df['category_id']==19),'category_name']="Travel and Events"
df.loc[(df['category_id']==20),'category_name']="Gaming"
df.loc[(df['category_id']==22),'category_name']="People and Blogs"
df.loc[(df['category_id']==23),'category_name']="Comedy"
df.loc[(df['category_id']==24),'category_name']="Entertainment"
df.loc[(df['category_id']==25),'category_name']="News and Politics"
df.loc[(df['category_id']==26),'category_name']="How to and Style"
df.loc[(df['category_id']==27),'category_name']="Education"
df.loc[(df['category_id']==28),'category_name']="Science and Technology"
df.loc[(df['category_id']==29),'category_name']="Non Profits and Activism"
df.loc[(df['category_id']==30),'category_name']="Movies"
df.loc[(df['category_id']==43),'category_name']="Shows"
df.head()
```

```python
df['year']=df['publish_time'].dt.year
yearly_count=df.groupby('year')['video_id'].count()
yearly_count.plot(kind='bar',xlabel="year",ylabel="Total publish Count",title="Total Publish Video Per View")
plt.show()
```

```python
import matplotlib.pyplot as plt
category_views = df.groupby('category_name')['views'].sum().reset_index()

top_categories = category_views.sort_values(by='views', ascending=False).head(5)

plt.pie(top_categories['views'], labels=top_categories['category_name'], autopct='%1.1f%%', startangle=140)
plt.title('Top 5 Categories by Total Views')
plt.axis('equal')
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(12, 6))
category_counts = df['category_name'].value_counts()
plt.bar(category_counts.index, category_counts.values)
plt.xticks(rotation=90)
plt.title('Video Count by Category', fontsize=20)
plt.xlabel('Category Name', fontsize=15)
plt.ylabel('Video Count', fontsize=15)

plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

videos_per_hour = df['publish_hour'].value_counts().sort_index()
heatmap_data = videos_per_hour.reset_index()
heatmap_data.columns = ['Hour of Day', 'Number of Videos']

plt.figure(figsize=(10, 6))
heatmap_plot = sns.heatmap(
    heatmap_data.set_index('Hour of Day').T, 
    cmap='YlGnBu',
    annot=True, 
    fmt='d', 
    cbar_kws={'label': 'Number of Videos'},  # Color bar label
)


plt.title('Number of Videos Published per Hour', fontsize=20)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df['publish_time'] = pd.to_datetime(df['publish_time'])
df['publish_date'] = df['publish_time'].dt.date
video_count_by_date = df['publish_date'].value_counts().sort_index()

plt.figure(figsize=(12, 6))
plt.plot(video_count_by_date.index, video_count_by_date.values, marker='o',color='red')  # Line plot with markers
plt.title('Number of Videos Published per Date', fontsize=20)
plt.xlabel('Publish Date')
plt.ylabel('Number of Videos')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
sns.scatterplot(data=df, x='views', y='likes',color='purple')
plt.title("Views VS Likes",fontsize=30)
plt.xlabel("Views")
plt.ylabel("Likes")
plt.show()
```

```python
plt.figure(figsize=(14,8))
plt.subplots_adjust(wspace=0.2,hspace=0.4,top=0.9)
plt.subplot(2,2,1)
g=sns.countplot(x='comments_disabled',data=df)
g.set_title('Comments Disabled',fontsize=20)

plt.subplot(2,2,2)
g1=sns.countplot(x='ratings_disabled',data=df)
g1.set_title('Ratings Disabled',fontsize=20)

plt.subplot(2,2,3)
g2=sns.countplot(x='video_error_or_removed',data=df)
g2.set_title('Video Error or Removed ',fontsize=20)
```

```python
numeric_df = df.select_dtypes(include=['int64', 'float64'])

corr_matrix = numeric_df.corr()

import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 8))
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```
